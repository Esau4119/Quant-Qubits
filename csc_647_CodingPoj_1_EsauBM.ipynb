```python
# import qiskit and numpy
from qiskit import QuantumCircuit, visualization
from qiskit import Aer, BasicAer, assemble, transpile, execute
from qiskit.quantum_info import Statevector, Operator
from datetime import datetime
from qiskit.tools.visualization import plot_histogram, plot_state_city,plot_bloch_multivector
import qiskit.quantum_info as qi
import numpy as np
```

```python
# Circuit 1
# Single qubit initialized to the |0> state
QC1 = QuantumCircuit(1)

#Measurement Gate        
QC1.measure_all()

#Running Circuit
openSims = Aer.get_backend('statevector_simulator')
QC1 = transpile(QC1, openSims)

res = openSims.run(QC1)


# Grabbing/printing saved state vector
stateVec = res.result().get_statevector()
print("State Vector:", stateVec)

plot_bloch_multivector(stateVec, title = "Circuit 1")

# Question about Circuit #1: 
#  looking at the Bloch Sphere plot of the quantum state,
#   what is the observed/measured value of the quantum 
#   state for this single-qubit circuit? 


# Answer: 
#  The measured value of this single qubit would be 0 
#  because we have not applied any transformations yet. As well as 
#  default initialization to 0.
```

```python
# Circuit 2
QC2 = QuantumCircuit(1)

#x gate flips qubit
QC2.x(0)

#using 'aer_simulator_statevector' we need to manually save state vector
QC2.save_statevector()


# Running circuit
QC2 = transpile(QC2, openSims)
os = Aer.get_backend('aer_simulator_statevector')
res = os.run(QC2)

# Grabbing/printing saved state vector
stateVec = res.result().get_statevector()
print("State Vector:", stateVec)

#Obtaining values measured and printing them. 
value = res.result().get_counts()
print("\nValue measured:", list(value.keys())[0])

# bloch Sphere
plot_bloch_multivector(stateVec, title = "Circuit 2")
```
